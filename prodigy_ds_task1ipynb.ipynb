```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
files=pd.read_csv("/jobs_in_data_2024.csv",on_bad_lines='skip')
```

```python
files.head()
```

```python
df=files[['job_title','salary',]]
```

```python
df2=df.groupby(['job_title'])['salary'].mean()
```

```python
df2=df2.head(5)
```

```python
plt.figure(figsize=(10,6))
plt.bar(df2.index,df2.values,color='skyblue')
plt.title('Mean salary of different jobs')
plt.xlabel('job title')
plt.ylabel("Mean salary")
plt.tight_layout()
plt.show()
```
